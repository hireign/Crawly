```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
rootURL = "https://www.hotelhalifax.ca/"
# Ads http:// to the given URL because it is the only way to check for server response
# If the user will add to the URL directions then they will be deleted
# Example: 'https://moz.com/learn/seo/external-link' will turn to 'https://moz.com/'
# https://stackoverflow.com/questions/32314304/check-if-an-url-is-relative-to-another-ie-they-are-on-the-same-host
if rootURL.split('//')[0] != 'http:' and rootURL.split('//')[0] != 'https:':
    rootURL = 'http://' + rootURL
    #try adding https as well
rootURL = rootURL.split('/')[0] + '//' + rootURL.split('/')[2]    
source_code = requests.get(rootURL)
soup = BeautifulSoup(source_code.text, "lxml")
rootTitle = soup.find("title").text
rootURL
```

```python
import re
domainName = re.compile(r"https?://(www\.)?")
domainName = domainName.sub('', rootURL).strip().strip('/')
# domainName = domainName.split('.')[0]
# domainName
```

```python
alternativeURLs = []
alternativeURLs.append(domainName)
alternativeURLs.append("http://"+domainName)
alternativeURLs.append("http://"+domainName+"/")
alternativeURLs.append("http://"+domainName+"//")
alternativeURLs.append("https://"+domainName)
alternativeURLs.append("https://"+domainName+"/")
alternativeURLs.append("https://"+domainName+"//")
alternativeURLs.append("http://www."+domainName)
alternativeURLs.append("http://www."+domainName+"/")
alternativeURLs.append("http://www."+domainName+"//")
alternativeURLs.append("https://www."+domainName)
alternativeURLs.append("https://www."+domainName+"/")
alternativeURLs.append("https://www."+domainName+"//")
alternativeURLs
```

```python
externalURLKeywords = ["http", "www", ".com"]
exceptions = ["#", " ", "tel:", "callto:", "mailto:", "wp-content", ".pdf", "twitter.com", "facebook.com", "google.com", "pinterest.com", ".jpg", ".jpeg", ".webm"]
```

```python
def urlCrawler(url, current_page, max_pages, title):
    tempURL = ""
    tempTargetList = []
    print("now crawling page: "+url)
    current_page+=1
    if(current_page == 1):
        urlList.append(url)
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text)
    a = soup.findAll("a")
    outgoing_links = 0
    titleList.append(soup.find("title").text)
    for eachURL in a:
        href = eachURL.get("href")
        if(eachURL.get("title")):
            anchor = eachURL.get("title")
            anchor = anchor.replace("\n", "").strip()
        else:
            anchor = eachURL.text
            anchor = anchor.replace("\n", "").strip()
        if(href):
            if not any(x in href for x in exceptions):
                if (href not in alternativeURLs and rootURL+""+href not in alternativeURLs):
                    if(domainName in href):
                        tempURL = href
                    elif(len(href)>1 and href[0] == "/"):
                        tempURL = rootURL+""+href
                    elif any(x in href for x in externalURLKeywords):
                        tempURL = ""
                    else:
                        tempURL = rootURL+"/"+href
                if(tempURL!="" and tempURL not in tempTargetList):
                    sourceList.append(url)
                    targetList.append(tempURL)
                    anchorList.append(anchor)
                    tempTargetList.append(tempURL)
                    outgoing_links += 1
                    print("appending link: "+tempURL)
                    print("link anchor: "+anchor)
                    print("outgoing link count: "+str(outgoing_links))
                    if (tempURL not in urlList):
                        urlList.append(tempURL)
    if(current_page < max_pages and current_page < len(urlList)): 
        outgoingLinksList.append(outgoing_links)
        urlCrawler(urlList[current_page], current_page, max_pages, rootTitle)
    else:
        outgoingLinksList.append(outgoing_links)
```

```python
sourceList = []
targetList = []
anchorList = []
urlList = []
titleList = []
clickDepthList = []
outgoingLinksList = []
outgoing_links = 0
totalPages = 0
urlCrawler(rootURL, 0, 40, rootTitle)
```

```python
Links = {}
Links['source'] = sourceList
Links['target'] = targetList
Links['anchor'] = anchorList
linkInfoDF = pd.DataFrame(Links)
linkInfoDF.head(1000)
```

```python
pages = {}
pages['url'] = urlList
pages['title'] = titleList
pages['outgoing_links'] = outgoingLinksList
pageInfoDF = pd.DataFrame(pages)
pageInfoDF.head(1000)
```

```python
urlName = rootURL
urlName = urlName.replace("http://","")
urlName = urlName.replace("https://","")
urlName = urlName.replace("www.","")
# urlName = urlName.replace("/","")
linkInfoDF.to_csv(r'C:\Users\hiren\crawly\links_'+urlName+'.csv')
pageInfoDF.to_csv(r'C:\Users\hiren\crawly\pages_'+urlName+'.csv')
```
